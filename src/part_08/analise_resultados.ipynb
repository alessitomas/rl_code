# Análise da curva de aprendizado para o problema do CartPole

```python
import pandas as pd
df = pd.read_csv('results/statistics_cartpole.csv')
df.head()
```

```python
import seaborn as sns
sns.lineplot(data=df, x="episode", y="rewards").set(title="Reward por episódios")
```

```python
sns.lineplot(data=df, x="episode", y=df['rewards'].ewm(span=100).mean()).set(title="Reward por episódios (valor médio considerando janela de 100 episódios)")
```

# Análise da curva de aprendizado para o problema do LunarLander

```python
import pandas as pd
df = pd.read_csv('results/statistics_lunarlander.csv')
df.head()
```

```python
import seaborn as sns
sns.lineplot(data=df, x="episode", y="rewards").set(title="Reward por episódios")
```

```python
import matplotlib.pyplot as plt

sns.lineplot(data=df, x="episode", y=df['rewards'].ewm(span=100).mean()).set(title="Reward por episódios (valor médio considerando janela de 100 episódios)")
plt.axhline(200, color='red')
plt.show()
```
